```python
import numpy as np

class score_filter:
    def __init__(self, min_score : float, max_score : float):
        self.min_score = min_score
        self.max_score = max_score
    
    def fit(self, x, y, arr):
#         x = np.cumsum(np.array(x))
#         y = np.cumsum(np.array(y))
        for i in range(1, len(x)):
            x[i] += x[i - 1]
        for i in range(1, len(y)):
            y[i] += y[i - 1]
        self.pairs = []
        denumerator = x[-1] * y[-1]
        
        for i, j in arr:
            numerator = x[i] * y[j]
            score = float(numerator) / float(denumerator)
            if self.min_score < score and score < self.max_score:
                self.pairs.append((i, j))
        
    def __iter__(self):
        i = 0
        while i < len(self.pairs):
            yield self.pairs[i]
            i += 1
```

```python
a = score_filter(0, 1)
a.fit([1, 2, 3], [3, 2, 1], [(0, 0), (1, 1), (2, 2)])

it = iter(a)
for i in it:
    print(i)
```
